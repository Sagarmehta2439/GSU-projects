```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
import itertools
```

```python
# our search term, using syntax for Twitter's Advanced Search
search = '"elections"'

# the scraped tweets, this is a generator
scraped_tweets = sntwitter.TwitterSearchScraper(search).get_items()

# slicing the generator to keep only the first 100 tweets
sliced_scraped_tweets = itertools.islice(scraped_tweets, 10)

# convert to a DataFrame and keep only relevant columns
df = pd.DataFrame(sliced_scraped_tweets)[['date', 'content']]
df
```

```python

df_city = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
    'election near:"Atlanta" within:10km').get_items(), 10000))[['content']]
df_city.head()
```

```python
df_city['Text_length'] = df_city.content.str.split().str.len()
```

```python
df_city.sample(10)
```

```python
df = df_city[df_city['Text_length']>53]
df.shape
```

```python
df.reset_index(inplace=True)
```

```python
df = df['content']
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_diego = pd.read_excel('label.xlsx',engine='openpyxl')
df_maiya = pd.read_csv('Election_Atlanta_M.csv',encoding='latin')
df_diego
```

```python
df_diego = df_diego.iloc[1:,:3].rename(columns = {'Election_Atlanta':'Text'})
df_diego.head()
```

```python
df_maiya.head()
```

```python
df = pd.merge(df_diego,df_maiya,left_on='Unnamed: 0',right_on='index')
df.head()
```

```python
df.Label.value_counts()
```

```python
df.Label = np.where(df.Label == 'not negative',0,1)
```

```python
df.label = np.where(df.label == 0.0,0,1)
```

```python
labels_not_equal = np.where(df.Label != df.label)
```

```python
df_label = df.loc[labels_not_equal,'Text_x']
df_label = pd.DataFrame(df_label)
```

```python
df_label.head()
```

```python
list_of_notnegative_index = [4,16,37,48,54,55,57,58,61,66,67,79,82,83,87,90,94,96,102,104,106]
list_negative_index = [3,8,15,42,44,53,60,84,86,89,97,107,114,118]
```

```python
df.head()
```

```python
df_final = df[['Text_x','label']]
df_final.loc[list_of_notnegative_index,'label'] = 0
df_final.loc[list_negative_index,'label'] = 1
df_final.label.value_counts()
```

```python
df_final.to_csv('final_data.csv')
```
