### **1. Import the required packages and read the data.**

```python
## Write your code in Here
import numpy as np
import pandas as pd
customers = pd.read_excel("Customers.xls")
print("Customers--------------------")
print(customers.head())
print("\n\nItems Ordered-------------------------")
items_ordered = pd.read_excel('ItemsOrdered.xls')
print(items_ordered.head())
st = pd.read_excel('SalesTerritory.xls')
print("\nSalesTerritory--------------------------")
print(st.head())
```

### **2. Produce a list of customers we have in North Carolina, and show how many there are.**

```python
## Write your code in Here
cust_state_NC = customers[customers['StateName'] == 'North Carolina']
cust_state_NC[['CustomerID','FirstName','LastName','StateName']]
print(f"The count of customers from North Carolina are \n{cust_state_NC.value_counts()}")
```

### **3. For each of the items ordered by a customer on a day, show the total price (sometimes they ordered more than 1 item).**

```python
## Write your code in Here
cust_info = pd.merge(customers,items_ordered,on = 'CustomerID')
cust = (cust_info
       .assign(Total_Price = cust_info.Quantity* cust_info.Price)
       .groupby(['CustomerID','OrderDate']).sum())[['Quantity','Price','Total_Price']]
cust

```

### **4. Show a list of customers, and the total amount of money they have spent. Show the highest spenders to appear first.**

```python
## Write your code in Here
cust_info = pd.merge(customers,items_ordered,on = 'CustomerID')
#cust_info.groupby(['CustomerID'])['Total_Price'].sum()
cust_info_2= cust_info.assign(Total_Price = cust_info.Quantity * cust_info.Price)
cust_info_2.head()
cust_spend = cust_info_2.groupby(['CustomerID'])['Total_Price'].agg(['sum']).sort_values(by = ['sum'],ascending = False)
cust_spend
```
