### **Import the required packages and read the data.**

```python
## Write your code in Here
import pandas as pd
customers = pd.read_excel('Customers.xls')
employees = pd.read_excel('Employees.xls')
items = pd.read_excel('ItemsOrdered.xls')
sales= pd.read_excel('SalesTerritory.xls')
excel_files = [customers,employees,items,sales]
for file in excel_files:
    print(f"------------------------------------------------\n{file.head()}")
```

### Question 1
**Write a function called `get_manager` that takes as its one argument the Pandas DataFrame "Employees" and returns a DataFrame containing all employees (EmployeeID, first name, middle name, last name), and their manager's first and last name. The columns in the output DataFrame should be: EmployeeID, FirstName, MiddleName, LastName, ManagerFirstName, ManagerLastName.**



```python
def get_manager(employees):
    df = employees.merge(employees,left_on = "ManagerID",right_on = 'EmployeeID')
    df = df.rename(columns = {'EmployeeID_x':'EmployeeID','FirstName_x':'FirstName','MiddleName_x':'MiddleName','LastName_x':'LastName','FirstName_y':'ManagerFirstName','LastName_y':'ManagerLastName'})
    return df[['EmployeeID','FirstName','MiddleName', 'LastName','ManagerFirstName','ManagerLastName']]
get_manager(employees)
```

### Question 2
**Write a functon called `get_spend_by_order` that takes as its two arguments the Pandas DataFrames "Orders" and "Customers", and returns a DataFrame with the following columns: "FirstName", "LastName", "Item", "TotalSpent", listing all cutomer names, their purchased items, and the total amount spend on that item (remember that the "Price" listed in "Orders" is the _price per item_).**



```python
## Write your code in Here
def get_spend_by_order(df1,df2):
    df3 = pd.merge(df1,df2,on = 'CustomerID')
    return df3.assign(TotalSpent = df3.Quantity*df3.Price)[["FirstName", "LastName", "Item", "TotalSpent"]]
    
get_spend_by_order(items,customers)   
```

### Question 3
**Write a function called `get_order_location` that takes three arguments: "Orders", "Customers", and "Territory", and returns a DataFrame containing the following columns:  "CustomerID", "Name", and "TotalItems", that gives, for each order, the  CustomerID, the name of the territory where the order was placed, and the total number of items ordered (yes, 2 ski poles counts as 2 items).** 



```python
## Write your code in Here
def get_order_location(df1,df2,df3):
    df4 = pd.merge(df1,df2,on = 'CustomerID')
    df5 = pd.merge(df4,df3,left_on ='SalesTerritoryID',right_on = 'TerritoryID')
    return df5.groupby(['CustomerID','Name'])[['Quantity']].sum().rename(columns = {'Quantity':"TotalItems"})
    
get_order_location(items,customers,sales)
```

### Question 4
**Write a function called `employee_info` that takes one argument: "Employees", and returns a DataFrame containing the following columns: JobTitle, NumberOfEmployees, and MeanVacationHours, containing all job titles, the number of employees with that job title, and the mean number of vacation days for employees with that job title.** 

```python
## Write your code in Here
def employee_info(df):
    return df.groupby(['JobTitle']).agg({'EmployeeID':'count','VacationHours':'mean'}).rename(columns = {'EmployeeID':'NumberOfEmployees','VacationHours':'MeanVacationHours'})
    
employee_info(employees)
```

### Question 5 (Bonus --25 points)
**Show a list of customers with the total amount of money they have spent, along with the territory they are in, sorted by total spending in descending order.** 

```python
## Write your code in Here
def cust_spend_by_territory(df1,df2,df3):
    df4 = pd.merge(df1,df2,on = 'CustomerID')
    df5 = pd.merge(df4,df3,left_on ='SalesTerritoryID',right_on = 'TerritoryID')
    return df5.assign(TotalSpent = df5.Quantity*df5.Price).groupby(['CustomerID','Name','TerritoryID']).sum().sort_values(by = 'TotalSpent',ascending = False)[['TotalSpent']]

cust_spend_by_territory(items,customers,sales)     
```
